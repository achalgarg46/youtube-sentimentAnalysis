```python
# =============================================================================
# Get Comments of Top Videos
# =============================================================================
video_id_pop = []
channel_pop = []
video_title_pop = []
video_desc_pop = []
comments_pop = []
comment_id_pop = []
reply_count_pop = []
like_count_pop = []

from tqdm import tqdm
for i, video in enumerate(tqdm(video_id, ncols = 100)):
    response = service.commentThreads().list(
                    part = 'snippet',
                    videoId = video,
                    maxResults = 100, # Only take top 100 comments...
                    order = 'relevance', #... ranked on relevance
                    textFormat = 'plainText',
                    ).execute()
    
    comments_temp = []
    comment_id_temp = []
    reply_count_temp = []
    like_count_temp = []
    dislike_count_temp = []
    for item in response['items']:
        comments_temp.append(item['snippet']['topLevelComment']['snippet']['textDisplay'])
        comment_id_temp.append(item['snippet']['topLevelComment']['id'])
        reply_count_temp.append(item['snippet']['totalReplyCount'])
        like_count_temp.append(item['snippet']['topLevelComment']['snippet']['likeCount'])

    comments_pop.extend(comments_temp)
    comment_id_pop.extend(comment_id_temp)
    reply_count_pop.extend(reply_count_temp)
    like_count_pop.extend(like_count_temp)

    video_id_pop.extend([video_id[i]]*len(comments_temp))
    channel_pop.extend([channel[i]]*len(comments_temp))
    video_title_pop.extend([video_title[i]]*len(comments_temp))
    video_desc_pop.extend([video_desc[i]]*len(comments_temp))
    
query_pop = [query] * len(video_id_pop)
```

# Removing Emoji's and further data cleaning

```python
import re

def Clean_Text(text):
    """
    text: input for the text
    Disclaimer:
    This is a ongoing cleaning function
    """

    text = text.lower()
    text = re.sub(r"what's", "what is ", text)
    text = re.sub(r"\'s", " ", text)
    text = re.sub(r"\'ve", " have ", text)
    text = re.sub(r"can't", "can not ", text)
    text = re.sub(r"n't", " not ", text)
    text = re.sub(r"i'm", "i am ", text)
    text = re.sub(r"he's","he is ",text)
    text = re.sub(r"she's","she is ",text)
    text = re.sub(r"\'re", " are ", text)
    text = re.sub(r"\'d", " would ", text)
    text = re.sub(r"\'ll", " will ", text)
    text = re.sub(r"\'scuse", " excuse ", text)
    text = re.sub('\W', ' ', text)
    text = re.sub('\s+', ' ', text)
    text = re.sub(r'[^\w\s]','',text)
    text = text.strip(' ')
    return text
```

# Calc. polarity of all sentences/words

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import pandas as pd
sia = SentimentIntensityAnalyzer()
neutral = 0
positive = 0
negative = 0
raw_neutral = 0
raw_positive = 0
raw_negative = 0
sentiments = []
len(cleaned_comments)

def pct_score(neu,neg, pos):
    total = neu+pos+neg
    #print(total)
    pct_neu = 100 * float(neu/total)
    pct_neg = 100 * float(neg/total)
    pct_pos = 100 * float(pos/total)
    #print(pct_neu, pct_neg, pct_pos)
    return(pct_neu, pct_neg, pct_pos)

for i in range(0,len(cleaned_comments)):
            comment = cleaned_comments[i]
            score = sia.polarity_scores(comment)
            sentiments.append([score['neg'], score['pos'], score['neu'], score['compound']])
            #print(comment,score)
            if score['compound'] >= 0.2 : 
                positive = positive +1
            elif score['compound'] <= - 0.1 :
                negative = negative +1
            else : 
                neutral = neutral+1

pct_neutral, pct_negative, pct_positive = pct_score(neutral,negative, positive)
                
for i in range(0,len(raw_comments)):
            comment = raw_comments[i]
            score = sia.polarity_scores(comment)
            sentiments.append([score['neg'], score['pos'], score['neu'], score['compound']])
            #print(comment,score)
            if score['compound'] >= 0.2 : 
                raw_positive = raw_positive +1
            elif score['compound'] <= - 0.05 :
                raw_negative = raw_negative +1
            else : 
                raw_neutral = raw_neutral+1

pct_Rneutral, pct_Rnegative, pct_Rpositive = pct_score(raw_neutral,raw_negative,raw_positive)

                
#print("Before cleaning the data -> We have ", raw_neutral, "Neutral commments,", raw_negative, "Negative comments, and ",raw_positive," Positive comments")
#print("After cleaning the data -> We have ", neutral, "Neutral commments,", negative, "Negative comments, and ",positive," Positive comments")
print("After cleaning the data -> We have ", pct_neutral, "% Neutral commments,", pct_negative, "% Negative comments, and ",pct_positive," % Positive comments")
print("Before cleaning the data -> We have ", pct_Rneutral, "% Neutral commments,", pct_Rnegative, "% Negative comments, and ",pct_Rpositive," % Positive comments")

df_test[['neg','neu','pos','compound']] = pd.DataFrame(sentiments)
df_test.head()
```
